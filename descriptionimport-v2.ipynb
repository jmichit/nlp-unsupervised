```python
season=3
file = 'Star Trek_ The Next Generation - Episode Guide - Season {}.html'.format(season)

with open('data/descriptions/'+file, 'r') as f:
    response = f.read()
bs = BeautifulSoup(response)
```

```python
rs = bs.find_all('p')
```

```python
ll = []

for p in rs[2:]:
    dd ={}
    dd['season'] = season
    dd['summary'] = p.text
    #print(p)
    ll.append(pd.DataFrame(dd, index=[0]))
df = pd.concat(ll)
```

```python
df
```

```python
#df.reset_index(drop=True, inplace=True)
#df.iloc[12,:] = None
#df.dropna(inplace=True)
```

```python
df['ep'] = df.summary.apply(lambda x:x.split('.')[0])
df['summary2'] = df.summary.apply(lambda x:x.split('.')[1])
df['title'] = df.summary2.apply(lambda x:x.split(chr(8211))[0])
df['summary3'] = df.summary2.apply(lambda x:x.split(chr(8211))[1])
df.drop(columns=['summary', 'summary2'], inplace=True)
df.rename(columns={'summary3':'summary'}, inplace=True)
```
